# Librerías y carga del fichero

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import seaborn as sns
import cufflinks as cf
import plotly.express as px
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy.stats import f_oneway

plt.rcParams["figure.figsize"] = (8,8)
```

```python
df = pd.read_csv('pisos_proyecto5.csv',encoding='latin-1')
df.head()
```

```python
pd.options.plotting.backend = 'plotly'
```

# Los precios siguen una normal?

```python
df['Precio'].iplot(kind='hist')
```

```python
stats.shapiro(df['Precio'])
```

p-value es menor de 0,05 por lo que, como intuíamos por el gráfico, no sigue una distribución normal

# Hay diferencias de precios por distritos?

```python
df[['Precio', 'Distrito']].pivot(columns='Distrito', values='Precio').iplot(kind='box')
```

```python
alta = df.query('Zona_cat in ["Alta"]')['Precio']
popular = df.query('Zona_cat in ["Popular"]')['Precio']
```

```python
sns.distplot(alta)
sns.distplot(popular)
```

```python
df[['Zona_cat', 'Precio']].pivot(columns='Zona_cat', values='Precio').iplot(kind='box')
```

```python
stats.mannwhitneyu(alta, popular)
```

Zona definida como alta se confirma más cara que la zona definida como popular

# Los tamaño en m2 son diferentes por las zona definidas?

```python
df[['Zona_cat', 'M2']].pivot(columns='Zona_cat', values='M2').iplot(kind='box')
```

```python
alta_m2 = df.query('Zona_cat in ["Alta"]')['M2']
popular_m2 = df.query('Zona_cat in ["Popular"]')['M2']
```

```python
f_oneway(alta_m2, popular_m2)
```

```python
stats.kruskal(alta_m2, popular_m2)
```

Zona definida como alta se confirma que tiene precios más grandes que la zona definida como popular

# Scatter interactivo relación x-y

```python
color = df[['Zona', 'Distrito', 'Barrio']]

variables = df[['Precio', 'M2', 'Habitaciones', 'Baños', 'Terraza',
       'Amueblado', 'Parking', 'Bus', 'Metro_y_Ferrocarril', 'RENFE',
       'Tramvia', 'Gasto_mensual_familiar', 'Ingreso_mensual_familiar',
       'Tasa_paro', 'Num_viviendas', 'Superficie_media_viviendas',
       'Nacionalidad_española', 'Nacionalidad_extranjera', 'Población',
       'Hombres', 'Mujeres']]
```

```python
@interact(x=variables, y=variables, color=color)
def explore_flats(x,y, color):
    sns.relplot(data=df, x=x, y=y, hue=color)
```

# Funcion para saber que equipo gano y relacion de goles a favor

```python
def resultados(s):
    for i in range(len(s)):
        if s[i]=='H':
            s[i]='gano'
        elif s[i]=='A':
            s[i]='perdio'
        else:
            s[i]='empato'
    return s
        
resultado = epl2.FTR.tolist()


epl2['resultado']= resultados(resultado)
epl2
```

# Resultados positivos local

```python
gana_local = epl2[epl2['resultado']=='gano']
gana_local
```

```python
gana_local['goles a favor'] = epl2['FTHG']-epl2['FTAG']
gana_local
```

```python
gana_local2 = gana_local.groupby(['HomeTeam'], as_index=False)['FTHG','FTAG','goles a favor'].sum()

gana_local2
gana_local2.iplot(kind='bar', x='HomeTeam', xTitle='HomeTeam',
           yTitle='goles', title='equipos epl')
```
